# Creating Taxonomy Tree

```python
import csv
taxonomy_CSV = 'action-all-levels.csv'

opne_CSV = open( taxonomy_CSV, 'r' )
reader = csv.reader( opne_CSV )

taxonomy_tree = {'name':'Taxonomy Tree','children':list()}


def get_node(node, tree, colorCode):
    # adds the nodes if does not exist
    # returns the existing or added node
    children = tree['children']
    output_node = {'name': node,'children':list(), 'colorCode': colorCode}
    to_insert = True
    for child in children:
        if child['name'] == node:
            output_node = child
            to_insert = False
    if to_insert:
        children.append(output_node)
    return output_node


for line in reader:
    level1_node = get_node(line[0],taxonomy_tree, line[0])
    level2_node = get_node(line[1],level1_node, line[0])
    level3_node = get_node(line[2],level2_node, line[0])
    get_node(line[3],level3_node, line[0])

```
